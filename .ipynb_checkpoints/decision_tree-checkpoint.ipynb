# Два центра (Two_centers)

# Решение (desicion)

### Библиотеки

```python
from collections import defaultdict
```

### Чтение входных данных

```python
n = int(input())
edges = []
for _ in range(n - 1):
    u, v, w = map(int, input().split())
    edges.append((u, v, w))
print(edges)
```

### Инициализация графа

#### Функция для добавления ребра в граф

```python
def add_edge(graph, u, v, w):
    graph[u].append((v, w))
    graph[v].append((u, w))
```

```python
graph = defaultdict(list)

for u, v, w in edges:
    add_edge(graph, u, v, w)
    
print(graph)
```
